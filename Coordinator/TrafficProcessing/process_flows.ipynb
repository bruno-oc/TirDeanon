```python
import pickle
from process_flows import generate_dataset, RAW_PATH, PROCESSED_PATH
```

```python
# Process the Client flows
TOTAL_FILES = 300
client_data = []

for f in range(TOTAL_FILES):
    file_name = f"tir/client/cap{f}"
    if not os.path.exists(RAW_PATH + file_name + '-A-upstream.pcapng'):
        continue
    client_data += generate_dataset(file_name)
print(f'Total client flows {len(client_data)}')

pickle.dump(client_data, open(PROCESSED_PATH + 'tir/client_data.pkl', 'wb'))
```

```python
# Process the Peers flows
TOTAL_FILES = 200
peers_data = []

for f in range(TOTAL_FILES):
    file_name = f"tir/peers/cap{f}"
    if not os.path.exists(RAW_PATH + file_name + '-A-upstream.pcapng'):
        continue
    peers_data += generate_dataset(file_name)
print(f'Total peers flows {len(peers_data)}')

pickle.dump(peers_data, open(PROCESSED_PATH + 'tir/peers_data.pkl', 'wb'))
```
